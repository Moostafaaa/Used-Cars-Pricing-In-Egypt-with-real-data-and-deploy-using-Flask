```python
connection = sqlite3.connect("data_db.db")
cr = connection.cursor()

# what are the prices of top 10 most expensive cars?
q1 = cr.execute("SELECT price FROM data ORDER BY price DESC LIMIT 10").fetchall()

# what is the model of the top 10 most expensive cars?
q2 = cr.execute("SELECT Model FROM data ORDER BY price DESC LIMIT 10").fetchall()

# what is the mean price for each maker and each model ?
q3 = cr.execute("SELECT AVG(price) , Make FROM data GROUP BY Make").fetchall()
q4 = cr.execute("SELECT AVG(price) , Model FROM data GROUP BY Model").fetchall()

# what are the most 5 frequent colors of cars ?
q5 = cr.execute("SELECT COUNT(color) , color FROM data GROUP BY color ORDER BY COUNT(color) DESC LIMIT 5").fetchall()

# what is the most frequent color for most expensive cars?
q6 = cr.execute("SELECT COUNT(color) , color , price FROM data GROUP BY color ORDER BY price DESC").fetchall()

# what is the minimum price for each car model and year of use ?
q7 = cr.execute(" SELECT MIN(price), Model , year FROM data GROUP BY Model ORDER BY year DESC ").fetchall()

# what is the minimum price for each year of use and the car model ?
q8 = cr.execute(" SELECT MIN(price), year , Model FROM data GROUP BY year ORDER BY year DESC ").fetchall()

# what is the maximun price for each car model per year of use ?
q9 = cr.execute(" SELECT MAX(price), Model , year FROM data GROUP BY Model ").fetchall()


# what is the mean price for each model for each year of use ?
q10 = cr.execute(" SELECT AVG(price), Model , year FROM data GROUP BY Model ORDER BY year DESC ").fetchall()


# what are the minimun and maximum CC for each car maker?
q11 = cr.execute(" SELECT MIN(CC),MAX(CC) , Make FROM data GROUP BY Make ").fetchall()


connection.close()
```
